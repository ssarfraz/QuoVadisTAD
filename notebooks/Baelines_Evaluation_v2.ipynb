# Evaluate all Baselines and summarize

```python
datasets_to_eval = [GeneralDataset.MSL, GeneralDataset.UCR_1]
```

```python
from tad.eval_baselines import evaluate_datasets

def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
```

## Full Evaluation of our drawn baselines Methods on 3 Main datasets (WADI, SWAT and UCR)

```python
_, df_std = evaluate_datasets(
                      preprocessing="0-1",
                      eval_method='standard',
                      distance="euclidean",
                      skip_useless_datasets=False,
                      datasets_ordered=datasets_to_eval)
```

```python
df_std.to_pickle(Path(DEFAULT_RESOURCE_DIR, './results_dfs/baselines_using_standard.pkl'))
#pd.read_pickle(Path(module_path, './resources/results_dfs/baselines_using_standard.pkl'))
```

```python
df_std.style.apply(highlight_max, props='color:white;background-color:darkblue', axis=0)
```

```python
df_std.drop(['P', 'R','AUPRC'], axis=1, level=1).style.apply(highlight_max, props='color:white;background-color:darkblue', axis=0)
```
